```
!pip install opencv-python

import cv2


def optimize_video_gpu(input_path, output_path, target_fps=60, quality=23):
    # Buka video input
    cap = cv2.VideoCapture(input_path)

    # Dapatkan informasi video asli
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    frame_rate = int(cap.get(cv2.CAP_PROP_FPS))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Buat objek VideoWriter untuk video output dengan GPU
    fourcc = cv2.VideoWriter_fourcc(*'hevc')  # Menggunakan codec H.265 (HEVC)
    out = cv2.VideoWriter(output_path, fourcc, target_fps, (frame_width, frame_height), True)

    current_frame = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Tulis frame ke video output dengan GPU
        out.write(frame)

        current_frame += 1
        progress = (current_frame / total_frames) * 100
        print(f"Progress: {progress:.2f}%")

    # Tutup video streams
    cap.release()
    out.release()
    cv2.destroyAllWindows()

# Ganti 'input_video.mp4' dengan path video asli Anda
# Ganti 'output_video.mp4' dengan path tempat menyimpan video hasil optimasi
optimize_video_gpu('/content/drive/MyDrive/Torrent/FC2-PPV-3669732/hhd800.com@FC2-PPV-3669732.mp4', '/content/drive/MyDrive/Torrent/FC2-PPV-3669732/new.mp4', target_fps=60)
```
